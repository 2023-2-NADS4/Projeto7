# 1- Quantidade de restaurantes no Brasil por Estado e Porte e categoria

```python
df_quant = df.filter(["Estado","Porte","Categoria"])
# Criando uma nova coluna 'count' para contar as ocorrências
df_quant = df.filter(["Estado","Porte","Categoria"])
df_quant['count'] = 1
# Agrupando os dados por 'Estado', 'Porte' e 'Categoria' e somando a coluna 'count'
df_grouped = df_quant.groupby(["Estado","Porte","Categoria"]).sum().reset_index()
fig = px.treemap(df_grouped, path=["Estado", "Porte", "Categoria"], values='count')
fig.show()
```

Esse nosso gráfico treemap mostra a relação entre a quantidade de restaurantes no Brasil, distribuidos por estado, porte e categoria. Cada retângulo no gráfico representa um estado, e os retângulos menores dentro de cada estado representam diferentes portes e categorias de restaurantes.

A área de cada retângulo é proporcional à contagem de restaurantes daquela categoria e porte naquele estado. Portanto, quanto maior a área do retângulo, maior a quantidade de restaurantes daquela categoria e porte naquele estado.

Este gráfico é útil para entender a distribuição de diferentes tipos e tamanhos de restaurantes em cada estado. Ele pode ajudar a identificar quais estados têm mais restaurantes de um determinado porte ou categoria, o que pode ser útil para análises de mercado ou planejamento de negócios.

```python
df_quant = df.filter(["País","Estado","Cidade"])
estado_counts = df_quant['Estado'].value_counts()
fig = px.pie(estado_counts.reset_index(), values='Estado', names='index', title='Distribução de Restaurantes por Estado', hole=0.6)
fig.show()
```

Aqui uma versão mais simples (Nosso Comic Sans dos gráficos), apenas para mostrar a distribução dos restaurantes no Brasil por Estados.

#4- Funcionários X Garçons X Mesas:

```python
df_garcons = df.filter(["Quantidade de Funcionários","Quantidade de Garçons", "Quantidade de mesas", "Estados", "Porte"])
# Primeiro, vamos criar uma nova coluna 'Outros Funcionários' que é a diferença entre 'Quantidade de Funcionários' e 'Quantidade de Garçons'
df['Outros Funcionários'] = df['Quantidade de Funcionários'] - df['Quantidade de Garçons']
# Agora, vamos criar um novo DataFrame com as colunas que queremos
df_garcons = df[['Quantidade de Garçons', 'Outros Funcionários']]
total_garcons = df_garcons.sum()

fig = px.pie(total_garcons, values=total_garcons.values, names=total_garcons.index, title='Garçons VS Outros Funcionários', hole=0.6)
fig.show()
```

Nesse gráfico de rosca simples, podemos perceber que quase metade dos funcionarios do pais que trabalham em restaurantes, são garçons.

```python
df['Outros Funcionários'] = df['Quantidade de Funcionários'] - df['Quantidade de Garçons']
df_funcionarios = df[['Estado', 'Porte', 'Categoria', 'Quantidade de Garçons', 'Outros Funcionários']]

# Derrete o DataFrame para que cada linha seja um funcionário
df_funcionarios = df_funcionarios.melt(id_vars=['Estado', 'Porte', 'Categoria'], var_name='Tipo de Funcionário', value_name='Quantidade')

# Agrupa os dados por 'Estado', 'Porte', 'Categoria' e 'Tipo de Funcionário' e soma a coluna 'Quantidade'
df_grouped = df_funcionarios.groupby(['Estado', 'Porte', 'Categoria', 'Tipo de Funcionário']).sum().reset_index()

fig = px.treemap(df_grouped, path=['Estado', 'Porte', 'Categoria', 'Tipo de Funcionário'], values='Quantidade')
fig.show()
```

Agora em uma análise mais detalhada que consdera o Estado, o Porte e a Categoria do restaurante, podemos perceber que essa média práticamente se mantém.

```python
df_garcons = df.filter(["Quantidade de Garçons", "Quantidade de mesas"])
df_garcons['Razao'] = df_garcons['Quantidade de mesas'] / df_garcons['Quantidade de Garçons']
fig = px.scatter(df_garcons, x='Quantidade de mesas', y='Quantidade de Garçons', color='Quantidade de mesas', color_continuous_scale='bluered')
fig.update_layout(title_text='Quantidade de Mesas para Cada Garçom',
                  xaxis_title="Quantidade de Mesas",
                  yaxis_title="Quantidade de Garçons",
                  xaxis_nticks=36)
fig.show()
```

O código acima está gerando um gráfico de dispersão que mostra a relação entre a quantidade de mesas e a quantidade de garçons em cada restaurante. Cada ponto no gráfico representa um restaurante, com a posição do ponto indicando a quantidade de mesas (eixo x) e a quantidade de garçons (eixo y) naquele restaurante.

A cor de cada ponto é determinada pela quantidade de mesas no restaurante, com um gradiente de azul para vermelho indicando uma quantidade crescente de mesas.

Este gráfico é útil para entender a relação entre a quantidade de mesas e a quantidade de garçons em cada restaurante. Ele pode ajudar a identificar se os restaurantes estão adequadamente equipados com garçons em relação ao número de mesas que possuem.

Por exemplo, se um ponto está localizado mais à direita no gráfico, isso indica que o restaurante tem muitas mesas. Se esse ponto também estiver mais alto no gráfico, isso indica que o restaurante também tem muitos garçons, o que pode sugerir que o restaurante está adequadamente equipado para atender a um grande número de mesas. Por outro lado, se um ponto está mais à direita, mas mais baixo no gráfico, isso pode sugerir que o restaurante pode não ter garçons suficientes para atender adequadamente todas as suas mesas.

#5- Quantidade de funcionários com base no porte:

```python
df_funcionarios_porte = df.filter(["Quantidade de Funcionários", "Porte"])
# Calcular a média da quantidade de funcionários por porte
media_funcionarios = df_funcionarios_porte.groupby('Porte')['Quantidade de Funcionários'].mean().reset_index()
media_funcionarios  = media_funcionarios.sort_values('Quantidade de Funcionários')
# Criar o gráfico de barras com cores diferentes para cada porte
fig = px.bar(media_funcionarios, x='Porte', y='Quantidade de Funcionários', color='Quantidade de Funcionários', title='Média da Quantidade de Funcionários por Porte')
fig.show()
```

Nesse temos um gráfico bem simples que serve apenas para mostrar que atualmente, talvez pela falta de automação na área, quanto maior um restaurante, mais funcionários ele necessita.

#6- Ticket Médio por Estado:

```python
df_ticket_estado = df.filter(["Categoria","Ticket Médio", "Porte", "Estado"])

# Calcular a média do ticket por Estado
estado_media = df_ticket_estado.groupby('Estado')['Ticket Médio'].mean().reset_index()
estado_media = estado_media.sort_values('Ticket Médio')
# Criar o gráfico de barras para Estado
fig_estado = px.bar(estado_media, x='Estado', y='Ticket Médio', color='Estado', title='Média do Ticket por Estado')
fig_estado.show()
```

Aqui temos uma distribuição simples apenas para mostrar oquanto cada público gasta em média em suas refeições em cada estado do país. Sem discriminar a categoria ou o porte dos estabelecimentos.

#7- Ticket Médio por Porte e Categoria:

```python
df_ticket_porte = df.filter(["Categoria","Ticket Médio", "Porte", "Estado"])
# Calcula a média do 'Ticket Médio' para cada combinação de 'Porte' e 'Categoria'
df_grouped = df_ticket_porte.groupby(['Porte', 'Categoria'])['Ticket Médio'].mean().reset_index()

fig = px.treemap(df_grouped, path=['Porte', 'Categoria'], values='Ticket Médio', title='Média do Ticket por Porte e Categoria')
fig.show()
```

Agora deixamos de pensar na distribuição por estados e pensamos apenas na relação do ticket com o Porte e a Categoria do restaurante.

Aqui percebemos que os maiores tickets são sempre os de rodízios, independente do porte.

Percebemos também que exite uma variação entre bar e restaurante em cada porte.

E por ultimo vemos essa média se manter no caso das Padarias e Cafés.

```python
# Filtra os restaurantes localizados em São Paulo
df_sp = df[df['Estado'] == 'São Paulo']

df_ticket_porte = df_sp.filter(["Categoria","Ticket Médio", "Porte"])
# Calcula a média do 'Ticket Médio' para cada combinação de 'Porte' e 'Categoria'
df_grouped = df_ticket_porte.groupby(['Porte', 'Categoria'])['Ticket Médio'].mean().reset_index()

fig = px.treemap(df_grouped, path=['Porte', 'Categoria'], values='Ticket Médio', title='Média do Ticket por Porte e Categoria em São Paulo')
fig.show()
```

Nesse temos uma análise apenas de São Paulo, onde estão a maioria dos estabelecimento, e que percebemos que os valores se mantém.

```python
# Filtra os restaurantes localizados fora de São Paulo
df_fora_sp = df[df['Estado'] != 'São Paulo']

df_ticket_porte = df_fora_sp.filter(["Categoria","Ticket Médio", "Porte"])
# Calcula a média do 'Ticket Médio' para cada combinação de 'Porte' e 'Categoria'
df_grouped = df_ticket_porte.groupby(['Porte', 'Categoria'])['Ticket Médio'].mean().reset_index()

fig = px.treemap(df_grouped, path=['Porte', 'Categoria'], values='Ticket Médio', title='Média do Ticket por Porte e Categoria fora de São Paulo')
fig.show()
```

E por último, apenas como prova real, um gráfico levando em  consideração todos os estados, menos São Paulo, mostrando mais uma vez que os valores praticamente se mantém, com exceção dos restaurantes de pequeno porte, onde vemos uma mudança significativa no valor do ticket médio dos bares.

#8- Faturamento mensal com base nos estados:

```python
# Converte a coluna 'Faturamento Mensal' para um tipo numérico
df['Faturamento Mensal'] = pd.to_numeric(df['Faturamento Mensal'], errors='coerce')
df_faturamento_estados = df.filter(['Faturamento Mensal','Estado'])
# Calcular a média do faturamento mensal por estado
media_faturamento = df_faturamento_estados.groupby('Estado')['Faturamento Mensal'].mean().reset_index()
# Ordenar os valores em ordem decrescente
media_faturamento = media_faturamento.sort_values('Faturamento Mensal', ascending=False)
# Criar o gráfico de barras com cores diferentes para cada estado
fig = px.bar(media_faturamento, x='Estado', y='Faturamento Mensal', color='Estado', title='Média do Faturamento Mensal por Estado')
fig.show()
```

Aqui um gráfico simples de barras mostrando a diferença das médias de faturamento de cada estado que nos trás uma análise interesante:

O faturamento de São Paulo fica entre os mais baixos. Talvez pela alta quantidade de restaurantes de pequeno porte na análise.
